```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv("C:/Users/Hari/Desktop/Strive School/DS_Strive_HW/DS_oct_2021/Day8/8. Matplotlib/data/company_sales_data.csv")
data.head(5)

# Write your code here
fig= plt.figure(figsize=(10,6))
ax= fig.add_subplot(1,1,1)
#set x axis
xlabels = range(13)
ax.set_xlabel("Month number", fontsize=20)
ax.set_xticks(xlabels)
ax.set_xticklabels(xlabels, fontsize=20 )
#set y axis
ylabels = range(100000,600000,100000)
ax.set_ylabel("Total profit", fontsize=20)
ax.set_yticks(ylabels)
ax.set_yticklabels(ylabels, fontsize=20 )
ax.set_ylim(100000,500000)

ax.set_title("Company profit per month", fontsize=30)
ax.plot(data['month_number'],data['total_profit'])
plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
fig= plt.figure(figsize=(15,10))
ax= fig.add_subplot(1,1,1)

labels = ["Facecream","Facewash","Toothpaste","Bathingsoap","Shampoo","Moisturizer"]

ax.set_title("Sales Data", fontsize=20)
ax.pie([data["facecream"].sum(),data["facewash"].sum(),data["toothpaste"].sum(),\
        data["bathingsoap"].sum(),data["shampoo"].sum(),data["moisturizer"].sum(),], normalize = True, \
      labels = labels, autopct=lambda p: '{:.1f}%'.format(p),textprops={'fontsize': 18})

ax.legend(labels,loc='lower right',prop={'size': 20})

```
